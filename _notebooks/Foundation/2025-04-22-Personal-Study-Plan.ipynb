## Key Terms

- `Variables` store information under a single name so you can read or change that information later without having to rewrite the actual value each time.
---
- `Data abstraction` hides unnecessary complexity in a data structure, letting you work with the parts that matter while the lower‑level details stay under the hood.
---
- `Mathematical expressions` combine numbers, variables, and operators (+, –, ×, ÷, etc.) to produce a value; programmers rely on them for calculations and problem‑solving.
---
- `Strings` are sequences of characters—letters, digits, punctuation—that hold and manipulate text such as words, sentences, or whole paragraphs.
---
- `Booleans` have only two possible states, true or false, and drive decision‑making by signaling whether a condition is met.
---
- `Conditionals` run different blocks of code depending on whether a given test is true or false, steering the program’s flow.
---
- `Nested conditionals` place one conditional inside another, enabling multi‑level decisions where an inner test runs only if an outer test passes.
---
- `Iteration` repeats a section of code, usually with a loop, either a set number of times or until a condition becomes false.
---
- `Lists` collect multiple items—numbers, strings, even other lists—inside a single container, making it easy to store, access, update, and loop through related data.

# Big Idea 5: Impact of Computing  
*(AP Computer Science Principles – College Board)*

---

## Lesson 5.1 · Beneficial and Harmful Effects  
### Summary  
Computing innovations have dramatically transformed society. These changes have improved healthcare, communication, and productivity, but they also raise concerns around privacy, bias, and the ethical use of technology.

### Key Concepts  
- **Positive Impacts**: Computing automates tedious tasks, increases access to information, and facilitates real-time communication worldwide. For example, apps and devices support people with disabilities, AI aids in disease diagnosis, and search engines make education more accessible.  
- **Negative Impacts**: When misused or poorly designed, technology can reinforce discrimination, threaten privacy, and contribute to job loss through automation. Ethical design and thoughtful policy are needed to minimize these harms and ensure equitable access and outcomes.

---

## Lesson 5.2 · Digital Divide  
### Summary  
The digital divide refers to the gap between those who have sufficient access to computing resources and those who do not. This divide can occur globally, regionally, or even within a single community.

### Key Concepts  
- **Access Inequality**: The divide often results from differences in infrastructure, income, and education. While some people benefit from constant high-speed internet and modern devices, others may have little or no access to technology.  
- **Impact**: This disparity affects educational achievement, career opportunities, and civic participation. Individuals and communities on the less connected side of the divide may struggle to compete or stay informed in an increasingly digital world.  
- **Closing the Gap**: Solutions include improving infrastructure, offering affordable internet and devices, and expanding digital literacy programs. Addressing the divide is essential to promote equity in the information age.

---

## Lesson 5.3 · Bias in Computing  
### Summary  
Bias in computing arises when systems or algorithms behave in ways that unfairly favor or disadvantage certain groups of people. This typically occurs when algorithms are trained on datasets that reflect societal inequalities or lack representation.

### Key Concepts  
- **Data-Driven Bias**: If a dataset used to train a model lacks diversity or reflects existing discrimination, the model may replicate or worsen those biases.  
- **Design Bias**: Bias can also be introduced by the developers' assumptions, either consciously or unconsciously, during system design.  
- **Consequences**: Biased algorithms can influence major decisions in hiring, lending, law enforcement, and healthcare, potentially reinforcing injustice. Mitigation strategies include ensuring diverse datasets, applying fairness metrics, and increasing transparency and accountability in algorithm design.

---

## Lesson 5.4 · Crowdsourcing  
### Summary  
Crowdsourcing involves soliciting input, solutions, or labor from a large group of people, typically via the internet. It has become a powerful tool in science, business, and creative fields.

### Key Concepts  
- **Scale and Collaboration**: Crowdsourcing allows large-scale collaboration that would be difficult for a single individual or organization. It’s used in projects like citizen science, large datasets labeling, and idea generation for product development.  
- **Efficiency and Reach**: By engaging the public, organizations can solve problems faster and often more cheaply than using internal teams. It democratizes participation and encourages innovation.  
- **Challenges**: Despite its advantages, crowdsourcing can raise issues around data quality, participant motivation, and fair compensation. Ethical use requires transparency, clear communication, and respect for contributors’ input.

---

## Lesson 5.5 · Randomness  
### Summary  
Randomness is essential in computing for simulating real-world unpredictability, enhancing security, and adding variation to digital experiences such as games.

### Key Concepts  
- **Applications**: Random values are used in simulations to model unpredictable systems (like weather), in games to introduce variability (like dice rolls), and in cybersecurity to generate encryption keys.  
- **Pseudorandomness**: Since computers are deterministic, they use pseudorandom number generators (PRNGs)—algorithms that produce values that appear random but follow a pattern.  
- **Quality and Security**: The quality of randomness affects accuracy and fairness. In security contexts, weak randomness can lead to vulnerabilities, so strong cryptographic techniques are required to ensure data protection.

---

## Lesson 5.6 · Legal and Ethical Concerns  
### Summary  
The widespread use of computing requires adherence to legal standards and ethical guidelines to ensure technology is used responsibly and users' rights are protected.

### Key Concepts  
- **Legal Protections**: Laws regulate data privacy, intellectual property, and the ethical treatment of children online. These include requirements for consent, copyright protection, and fair use.  
- **Ethical Responsibilities**: Developers must consider the social impacts of their products. Responsible computing includes being transparent about data collection, minimizing harm, and ensuring inclusive design.  
- **Balancing Innovation and Rights**: As technology evolves, it's important to strike a balance between innovation and protecting the rights and dignity of individuals.

---

## Lesson 5.7 · Safe Computing  
### Summary  
Safe computing practices help individuals protect their devices, personal information, and digital identities from cyber threats and misuse.

### Key Concepts  
- **Security Practices**: Users are encouraged to use strong, unique passwords, enable two-factor authentication, and regularly update software to patch vulnerabilities.  
- **Threat Awareness**: Being cautious with unknown links or attachments, recognizing phishing attempts, and avoiding suspicious downloads are critical habits for staying safe.  
- **Data Protection**: Encrypting sensitive files and maintaining backups helps prevent loss or theft. Organizations also develop policies and training to promote safe computing across teams and systems.
