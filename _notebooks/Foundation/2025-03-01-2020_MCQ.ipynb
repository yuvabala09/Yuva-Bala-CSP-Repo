### **1.4: Identifying and Correcting Errors (5/7) 71%**
Errors in programming can be categorized into:
- **Syntax errors**: Issues with the language’s grammar that prevent code execution.
- **Runtime errors**: Errors that occur during execution, such as division by zero.
- **Logic errors**: Mistakes that cause incorrect output despite running successfully.
Debugging techniques include **print statements**, **step-by-step execution**, and **test cases**.

### **3.4: Strings (0/1) 0%**
A **string** is a sequence of characters. Key operations include:
- **Concatenation** (`"Hello" + " World"` → `"Hello World"`).
- **Substring extraction** (`str[0:3]` extracts part of a string).
- **String length calculation** (`len(str)`).
- **Indexing** (`str[i]` retrieves a specific character).

---

### **3.6: Conditionals (0/1) 0%**
Conditional statements allow decision-making in code:
- **If-else statements** execute different code blocks based on conditions.
- **Relational operators** (`==`, `!=`, `<`, `>`).
- **Boolean expressions** (`True`, `False`).
- **Logical operators** (`AND`, `OR`, `NOT`) to combine conditions.

---

### **3.7: Nested Conditionals (½) 50%**
Nested conditionals are **if statements inside if statements**, allowing multi-step decision-making.
- Example:
  ```python
  if age >= 18:
      if has_id:
          print("Access granted")
      else:
          print("ID required")
  else:
      print("Access denied")
  ```
Proper indentation is crucial to avoid logical errors.

The Internet is a **decentralized network** that enables global communication:

- **HTTP/HTTPS protocols** manage web requests.
- The Internet is designed for **scalability & fault tolerance**.
