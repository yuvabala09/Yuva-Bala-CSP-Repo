# Summer Machine Learning Roadmap (10 Weeks)

##  Week 1–2: Foundations of Machine Learning
- **Goal:** Understand ML basics, terminology, and core concepts.
- **Topics:**
  - What is Machine Learning?
  - Types of ML: Supervised, Unsupervised, Reinforcement Learning
  - Regression vs Classification
  - Training vs Testing Data
  - Overfitting & Underfitting
  - Bias-Variance Tradeoff
- **Resources:**
  - [Google ML Crash Course](https://developers.google.com/machine-learning/crash-course)
  - YouTube: StatQuest’s ML Playlist
  - Book: *Hands-On ML with Scikit-Learn, Keras & TensorFlow* (Intro chapters)

---

##  Week 3–4: Python for Data Science & ML Tools
- **Goal:** Get hands-on with Python libraries used in ML.
- **Topics:**
  - NumPy, Pandas (data wrangling)
  - Matplotlib, Seaborn (visualization)
  - Scikit-learn (modeling basics)
- **Mini-Project:** Titanic dataset classification (Kaggle)
- **Resources:**
  - Kaggle's Python and Pandas Courses
  - [Scikit-learn Tutorials](https://scikit-learn.org/stable/tutorial/index.html)

---

##  Week 5–6: Supervised Learning in Practice
- **Goal:** Learn to implement and tune models.
- **Topics:**
  - Linear & Logistic Regression
  - K-Nearest Neighbors
  - Decision Trees and Random Forests
  - Model Evaluation: Accuracy, Precision, Recall, F1, ROC
- **Mini-Project:** Predict housing prices (Boston dataset or California Housing)
- **Resources:**
  - Coursera: Andrew Ng’s ML course (Weeks 2–4)
  - Kaggle Notebooks for real-world datasets

---

##  Week 7: Unsupervised Learning
- **Goal:** Explore structure in unlabeled data.
- **Topics:**
  - Clustering: K-Means, Hierarchical
  - Dimensionality Reduction: PCA
- **Mini-Project:** Group customers based on purchasing behavior
- **Resources:**
  - YouTube: StatQuest on Clustering & PCA
  - Scikit-learn documentation

---

##  Week 8–9: Neural Networks & Deep Learning
- **Goal:** Learn the basics of deep learning and neural nets.
- **Topics:**
  - Artificial Neural Networks
  - Activation functions
  - Training with TensorFlow/Keras
  - Overfitting and regularization in neural networks
- **Mini-Project:** Digit classification using MNIST
- **Resources:**
  - TensorFlow Tutorials: [https://www.tensorflow.org/tutorials](https://www.tensorflow.org/tutorials)
  - Book: *Deep Learning with Python* by François Chollet

---

##  Week 10: Final Project
- **Goal:** Apply what you've learned in a custom ML project.
- **Project Ideas:**
  - **ML-Powered News Classifier**: Categorize articles into politics, sports, etc.
  - **Border Wait Time Predictor**: Predict delays at US-Mexico crossings (based on your Crosswise project)
  - **Sentiment Analysis Tool**: Analyze movie or product review polarity
- **Deliverables:**
  - Well-commented Jupyter Notebook
  - README with problem statement, dataset, methods, and results
  - GitHub repository or personal site for showcasing

---

##  End of Summer Outcomes
- Solid understanding of core ML algorithms and when to use them
- Comfort using Python and ML libraries (Pandas, Scikit-learn, TensorFlow)
- One polished project you can showcase on your LinkedIn/GitHub/Portfolio
