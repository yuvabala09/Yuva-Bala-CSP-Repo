# FRQ

### 1. Why is binary search more efficient for large data than linear search?
Binary search is more efficient because it cuts the search space in half with each step, reducing the time complexity to O(log n) compared to linear search’s O(n). This makes it significantly faster for large datasets.

### 2. What happens if the list isn’t sorted and you try to use binary search?
If the list isn't sorted, binary search won’t work correctly because it relies on the order of elements to eliminate half the data at each step. It may return the wrong result or fail to find the target entirely.

### 3. Could you use binary search in a video game leaderboard or streaming service search bar? Why or why not?
Yes, binary search can be used in a leaderboard if the scores are sorted, allowing fast lookups. However, a streaming search bar may require partial matches or fuzzy searching, so more advanced search algorithms or data structures like tries might be better suited.
